```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def circle(a,b,r):
    fig, ax = plt.subplots(figsize=(5,5))
    t = np.linspace(0, np.pi * 2, 1000)
    x = a + r * np.cos(t)
    y = b + r * np.sin(t)
    ax.axvline(a, color='grey', linewidth=.5, linestyle='-')
    ax.axhline(b, color='gray', linewidth=.5, linestyle='-')
    ax.plot(a,b, color='r', marker='o') # center point
    
    rect=mpl.patches.Rectangle((-.5,-.5),1,1, fill=False)
    ax.add_patch(rect)
    ax.axis('off')
    ax.plot(x, y)
```

```python
circle(0,0,.5)
```

```python
n = 100000
# radius = .5
x=np.random.uniform(-.5, .5, n)
y=np.random.uniform(-.5, .5, n)
```

```python
pi_est= 4 * (a < .25).sum() / n ##  4 * (a < .25).sum() / n =  (a < .25).sum() / n * 4
pi_est
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x[a < .25], y[a < .25], s=.1)
plt.scatter(x[a >= .25], y[a > .25], color='r', s=.1)
plt.title(f'n = {n:,} estimated pi = {pi_est}, delta = {np.abs(np.pi - pi_est)}')
```
