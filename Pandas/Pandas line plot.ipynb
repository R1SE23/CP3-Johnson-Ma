```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter # for custom number format on x and y axis
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
pd.set_option('display.max_rows', 10)
```

```python
oil=pd.read_csv('https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/OPEC_monthly_oilprice.csv', index_col='Date', parse_dates= True)
oil
```

```python
oil.plot(kind='line')
```

```python
ax = oil.plot(kind='line', title='OPEC oil price US$/barrel', color='maroon')
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
ax.axhline(oil['Oil'].max(), color='yellowgreen', linestyle='--', linewidth=.8)
```

```python
oil['Oil'].mean()
```

```python
ax = oil.loc['2008-01-31':, :].plot(kind='line', title='OPEC oil price US$/barrel',
                                  color='maroon', figsize=(10,5))
ax.set_xlabel('')
ax.set_ylabel('US$/barrel')
ax.axhline(oil.loc['2008-01-31':, 'Oil'].mean(), 
           color='yellowgreen', linestyle='--', linewidth=.8)
```
