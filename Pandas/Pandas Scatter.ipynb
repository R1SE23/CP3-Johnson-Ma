```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.head()
```

```python
df.plot.scatter(x='displ', y='cty')
```

```python
df.plot.scatter(x='displ', y='cty', grid=True, title='City miles per gallon vs. engine size (liters)');
```

```python
df.sample(5)
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue','r':'red'}),
                title='City miles per gallon vs. engine size (liters)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue','r':'red'}),
                s=df.displ *10,
                title='City miles per gallon vs. engine size (liters)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue','r':'red'}),
                s=df.cyl *10,
                alpha=.5,
                title='City miles per gallon vs. engine size (liters)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue','r':'red'}),
                s=df.cyl *10, ###   for this one cyl
                alpha=.5,
                title='City miles per gallon vs. engine size (liters)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue','r':'red'}),
                s=30,
                alpha=.5,
                ylim=(0,40),
                xlim=(0,10),
                title='City miles per gallon vs. engine size (liters)');
```
